```python
swe_1['Date'] = swe_1['Datetime'].dt.date
usgs_2['Date'] = usgs_2['Datetime'].dt.date
```

```python
dataset = {}

list_swe = [swe_1, swe_2, swe_3]

list_usgs = [usgs_1, usgs_2]

for ii, usgs_station in enumerate(list_usgs):
    
    for jj, swe_station in enumerate(list_swe):
        
        usgs_station['Date'] = usgs_station['Datetime'].dt.date
        
        swe_station['Date'] = swe_station['Datetime'].dt.date
        
        
        dataset[f'S{jj}_U{ii}'] = pd.merge(usgs_station, swe_station, on='Date')
    
        
    
```

```python
#Set start date and end date
StartDate= '2017-10-01'
EndDate= '2023-09-30'
```

```python
station_name=list(dataset.keys())
station_name
```

```python
station_list=list(['Pascoes (PSC) & Gauge 1', 'Upper Tyndall Creek (UTC) & Gauge 1', 'Casa Vieja Meadows (CSV) & Gauge 1',
                  'Pascoes (PSC) & Gauge 2', 'Upper Tyndall Creek (UTC) & Gauge 2', 'Casa Vieja Meadows (CSV) & Gauge 2'])
```

```python
%%time

# Calculate the number of subplots needed based on the number of unique stations.
n_subplots = len(station_list)
# Determine the number of columns in the subplot grid by taking the ceiling of the square root of 'n_subplots'.
n_cols = int(math.ceil(math.sqrt(n_subplots)))
# Determine the number of rows in the subplot grid by dividing 'n_subplots' by 'n_cols' and taking the ceiling of that.
n_rows = int(math.ceil(n_subplots / n_cols))
# Set the figure size for the subplots.
figsize = (36, 20)
# Create a grid of subplots with specified number of rows and columns and figure size.
fig, axes = plt.subplots(n_rows, n_cols, figsize=figsize, dpi=300)
# Flatten the axes array for easier iteration.
# axes = axes.flatten()
if n_subplots > 1:
    axes = axes.flatten()
else:
    axes = [axes]
    
title_fontsize = 32
label_fontsize = 32
tick_fontsize = 28

# Iterate over the axes to plot the data for each station.
for i, ax in enumerate(axes):
    if i < n_subplots:
        # Extract the data for the current station from the dataset.
        station_name=list(dataset.keys())[i]
        temp_df_1 = dataset[station_name]
#         # Extract the data for the current station from the dataset.
#         temp_df_1 = dataset[dataset.station_id == station_list[i]]
#         temp_df_1 = Dataframe_dict[station_name]
        temp_df_1['Date'] = pd.to_datetime(temp_df_1['Date'])
        # Set 'datetime' as the index for plotting.
        temp_df_2 = temp_df_1.set_index('Date')
        # Plot the 'flow_cfs' data on the primary y-axis.
        ax.plot(temp_df_2.index, temp_df_2['qualifiers'], color='blue')
        # Set the x-axis limits from the first to the last year of data.
        start_year = pd.to_datetime(StartDate)
        end_year = pd.to_datetime(EndDate)
#         start_year = pd.to_datetime(f'{temp_df_1.datetime.dt.year.min()}-10-01')
#         end_year = pd.to_datetime(f'{temp_df_1.datetime.dt.year.max()}-09-30')
        ax.set_xlim(start_year, end_year)
        # Rotate x-axis labels for better readability.
        labels = ax.get_xticklabels()
        ax.set_xticklabels(labels, rotation=45, fontsize=tick_fontsize)

        # Create a second y-axis for the precipitation data.
        ax2 = ax.twinx()
        # Plot the 'precip(mm)' data as a bar graph on the secondary y-axis.
        ax2.bar(temp_df_2.index, temp_df_2['Snow water equivalent'], label='Inverted', color='red', alpha=1, width=25)
        # Set the y-axis limits for precipitation, flipping the axis to make bars grow downward.
        ax2.set_ylim(max(temp_df_2['Snow water equivalent']) + 100, -5)
        # Set the label for the secondary y-axis.
        ax2.set_ylabel('Snow water equivalent (in)', fontsize=label_fontsize)
        # Define custom ticks for the secondary y-axis.
        ax2.set_yticks(np.arange(0, max(temp_df_2['Snow water equivalent']) + 40, 20))
        ax2.tick_params(axis='y', labelsize=tick_fontsize)

        # Set the title of the subplot to the station ID.
        ax.set_title(f'{station_list[i]}', fontsize=title_fontsize)
        # Set the x-axis label for subplots in the last row.
        if i // n_cols == n_rows - 1:
            ax.set_xlabel('Water year', fontsize=label_fontsize)
        if i // n_cols != n_rows - 1:
            ax.set_xticklabels([])  
            ax.set_yticks(np.arange(0, max(temp_df_2['qualifiers']) + 450, 100))
        # Set the y-axis label for subplots in the first column.
        if i % n_cols == 0:
            ax.set_ylabel('Streamflow (cfs)', fontsize=label_fontsize)
            ax.tick_params(axis='x', labelsize=tick_fontsize)
            ax.tick_params(axis='y', labelsize=tick_fontsize)
        if i % n_cols != 0:
            ax.set_yticks([])
        if i % n_cols < 2:
            ax2.set_yticks([])
            ax2.set_ylabel('')

    else:
        # Hide any unused axes.
        ax.axis('off')

# Adjust layout to prevent overlapping elements.
plt.tight_layout()

# Set horizontal distance between subplots
plt.subplots_adjust(wspace=0.1, hspace=0.2)  

# Uncomment the line below to save the figure to a file.
plt.savefig(f'Streamflow_SWE.png', dpi=300, pad_inches=0.2, facecolor= 'White')
# Display the plot.
plt.show()
```
